```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus =  ["you were born with potential",
"you were born with goodness and trust",
# "you were born with ideals and dreams",
# "you were born with greatness",
# "you were born with wings",
# "you are not meant for crawling, so don't",
"you have wings",
"learn to use them and fly"
]
```

```python
def tokenizer(text):
    PATTERN = r'[\s,\.\!\;\?\'\:]+'
    return re.split(PATTERN, text)
```

```python
tf_idf_vect = TfidfVectorizer(tokenizer=tokenizer, smooth_idf=True, norm='l1')
tf_idf_text = tf_idf_vect.fit_transform(corpus)
terms = tf_idf_vect.get_feature_names()
```

```python
df = pd.DataFrame.sparse.from_spmatrix(tf_idf_text).stack()
df = df.reset_index()
df.columns = ['idDoc', 'idTerm', 'tfidf']
df = df[df.tfidf > 0]
df['term'] = df['idTerm'].apply(lambda x: terms[x])
```

```python
df.sort_values(['idDoc', 'tfidf'], ascending=True, inplace=True)
```
